```python
import csv
f = open('cig.csv', 'rb')
reader = csv.reader(f)
tar = []
nic = []
for row in reader:
    nums = row[0].split()
    tar.append(float(nums[1]))
    nic.append(float(nums[2]))
print tar
print nic
    
f.close()
```
