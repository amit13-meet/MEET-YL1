```python
import csv
f = open('pokemon.csv', 'rb')
reader = csv.reader(f)
hei = []
wei = []
index = 0
for row in reader:
    if index != 0:
        nums = row[0].split()
        hei.append(float(nums[3]))
        wei.append(float(nums[4]))
    index += 1

print hei
print wei
    
f.close()
```

```python
from pylab import *
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Pokemon Height Vs. Weight')
ax.set_xlabel('Height (Meters)')
ax.set_ylabel('Weight (Kilograms)')
scatter(hei,wei)
fig.savefig("PokeProject")
```

This scatter plot describes the connection between Pokemons height and weight in the first generation.
Most Pokemon are light and short, mostly the common ones, but the legendary Pokemon are unusually heavy and tall compared to most others.
